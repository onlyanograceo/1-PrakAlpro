```python
# Program cek anagram

kata1 = input("Masukkan kata pertama: ")
kata2 = input("Masukkan kata kedua: ")

# Mengubah ke huruf kecil dan mengurutkan
if sorted(kata1.lower()) == sorted(kata2.lower()):
    print("Anagram")
else:
    print("Bukan anagram")
```

```python
# Program menghitung frekuensi kata

kalimat = input("Masukkan kalimat: ")
kata_cari = input("Masukkan kata yang dicari: ")

# Ubah ke huruf kecil dan pisahkan jadi list kata
kata_list = kalimat.lower().split()

# Hitung jumlah kemunculan
jumlah = kata_list.count(kata_cari.lower())

print(f"{kata_cari} ada {jumlah} buah")
```

```python
# Program menghapus spasi berlebih

kalimat = input("Masukkan kalimat: ")

# Menghapus spasi berlebih
hasil = " ".join(kalimat.split())

print("Output:", hasil)
```

```python
# Program mencari kata terpendek dan terpanjang

kalimat = input("Masukkan kalimat: ")

# Pisahkan kalimat menjadi kata-kata
kata_list = kalimat.split()

# Cari kata terpendek dan terpanjang
terpendek = min(kata_list, key=len)
terpanjang = max(kata_list, key=len)

print("Terpendek:", terpendek)
print("Terpanjang:", terpanjang)
```
