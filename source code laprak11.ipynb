```python
def tiga_terbaik(data):
    data_urut = sorted(data, reverse=True)  
    return data_urut[:3]  

angka = [10, 5, 8, 20, 3, 15]
hasil = tiga_terbaik(angka)

print("3 bilangan terbaik:", hasil)
```

```python
def hitung_rata_rata():
    data = []
    
    while True:
        user_input = input("Masukkan bilangan (ketik 'done' untuk selesai): ")
        
        if user_input.lower() == "done":
            break
        
        try:
            angka = float(user_input)
            data.append(angka)
        except ValueError:
            print("Input tidak valid, masukkan angka!")

    if len(data) > 0:
        rata_rata = sum(data) / len(data)
        print("Rata-rata:", rata_rata)
    else:
        print("Tidak ada data yang dimasukkan.")

hitung_rata_rata()
```

```python
def kata_unik(nama_file):
    try:
        with open(nama_file, 'r') as file:
            teks = file.read().lower() 
            
        for tanda in [".", ",", "!", "?", ":", ";", "\"", "'"]:
            teks = teks.replace(tanda, "")
        
        kata_list = teks.split()  
        unik = set(kata_list)
        
        print("Kata-kata unik:")
        for kata in sorted(unik):
            print(kata)
    
    except FileNotFoundError:
        print("File tidak ditemukan.")

nama_file = input("Masukkan nama file: ")
kata_unik(nama_file)
```
