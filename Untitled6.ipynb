```python
try:
    a = float(input("Masukkan angka pertama: "))
    b = float(input("Masukkan angka kedua: "))
    hasil = a + b
    print("Hasil penjumlahan:", hasil)
except ValueError:
    print("Input harus berupa angka!")
```

```python
try:
    angka = float(input("Masukkan sebuah angka: "))
    if angka > 0:
        print("Angka Positif")
    elif angka < 0:
        print("Angka Negatif")
    else:
        print("Angka Nol")
except ValueError:
    print("Input harus berupa angka!")
```

```python
try:
    bilangan = int(input("Masukkan sebuah bilangan bulat: "))
    
    if bilangan % 2 == 0:
        print("Bilangan Genap")
    else:
        print("Bilangan Ganjil")

except ValueError:
    print("Input harus berupa bilangan bulat!")
```

```python
try:
    angka = float(input("Masukkan sebuah angka: "))
    hasil = "Angka Positif" if angka > 0 else "Angka Negatif" if angka < 0 else "Angka Nol"
    print(hasil)
except ValueError:
    print("Input harus berupa angka!")
```

```python
try:
    bulan = int(input("Masukkan bulan (1-12): "))
    if bulan < 1 or bulan > 12:
        print("Bulan tidak valid!")
    elif bulan == 2:
        print("Jumlah hari: 29")  # Tahun 2020 adalah tahun kabisat
    elif bulan in [4, 6, 9, 11]:
        print("Jumlah hari: 30")
    else:
        print("Jumlah hari: 31")
except ValueError:
    print("Input harus berupa angka!")
```

```python
try:
    sisi1 = float(input("Masukkan sisi 1: "))
    sisi2 = float(input("Masukkan sisi 2: "))
    sisi3 = float(input("Masukkan sisi 3: "))
    if sisi1 <= 0 or sisi2 <= 0 or sisi3 <= 0:
        print("Input tidak valid! Sisi harus lebih dari 0.")
    elif sisi1 == sisi2 == sisi3:
        print("3 sisi sama")
    elif sisi1 == sisi2 or sisi1 == sisi3 or sisi2 == sisi3:
        print("2 sisi sama")
    else:
        print("Tidak ada yang sama")
except ValueError:
    print("Input harus berupa angka!")
```
