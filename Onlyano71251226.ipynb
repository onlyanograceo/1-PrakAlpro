```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
X = np.linspace(0,10)
y = np.sin(X)
z = np.cos(X)
plt.plot(X,y,'b',X,z,'r')
plt.xlabel('radians');
plt.ylabel('value');
plt.title('plotting demonstration')
plt.legend(['sin','cos'])
plt.grid()
```
