```python
def cek_angka(a, b, c):
    # Cek semua nilai berbeda
    berbeda = (a != b) and (a != c) and (b != c)
    
    # Cek apakah ada dua angka yang jika dijumlahkan sama dengan angka ketiga
    jumlah_sesuai = (a + b == c) or (a + c == b) or (b + c == a)
    
    # Mengembalikan True jika semua ketentuan terpenuhi
    return berbeda and jumlah_sesuai


# Contoh penggunaan
print(cek_angka(2, 3, 5))  # True
print(cek_angka(1, 2, 3))  # True
print(cek_angka(2, 2, 4))  # False
print(cek_angka(1, 2, 4))  # False
```

```python
def cek_digit_belakang(a, b, c):
    # Ambil digit paling kanan
    d1 = a % 10
    d2 = b % 10
    d3 = c % 10
    
    # Cek apakah minimal dua digit sama
    if d1 == d2 or d1 == d3 or d2 == d3:
        return True
    else:
        return False


# Membaca input dari pengguna
a = int(input("Masukkan bilangan pertama: "))
b = int(input("Masukkan bilangan kedua: "))
c = int(input("Masukkan bilangan ketiga: "))

# Menunjukan hasil
print(cek_digit_belakang(a, b, c))
```

```python
# Lambda function konversi suhu
c_to_f = lambda c: (9/5) * c + 32
c_to_r = lambda c: 0.8 * c

# Test case
# Input C = 100
print("C = 100 -> F =", c_to_f(100))  # Output: 212

# Input C = 80
print("C = 80 -> R =", c_to_r(80))    # Output: 64

# Input C = 0
print("C = 0 -> F =", c_to_f(0))      # Output: 32
```
