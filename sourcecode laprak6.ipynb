```python
def perkalian(a, b):
    hasil = 0
    for i in range(a):
        hasil = hasil + b
    return hasil

# input dari pengguna
x = int(input("Masukkan angka pertama: "))
y = int(input("Masukkan angka kedua: "))

# memanggil fungsi
hasil = perkalian(x, y)

print(x, "x", y, "=", hasil)
```

```python
def ganjil(bawah, atas):
    if bawah < atas:
        for i in range(bawah, atas + 1):
            if i % 2 != 0:
                print(i, end=", ")
    else:
        for i in range(bawah, atas - 1, -1):
            if i % 2 != 0:
                print(i, end=", ")

# input dari pengguna
bawah = int(input("Masukkan batas bawah: "))
atas = int(input("Masukkan batas atas: "))

# memanggil fungsi
ganjil(bawah, atas)
```

```python
jumlah_mk = int(input("Masukkan jumlah mata kuliah: "))
total_nilai = 0

for i in range(jumlah_mk):
    nilai = input("Masukkan nilai mata kuliah ke-" + str(i+1) + " (A/B/C/D): ").upper()
    
    if nilai == "A":
        bobot = 4
    elif nilai == "B":
        bobot = 3
    elif nilai == "C":
        bobot = 2
    elif nilai == "D":
        bobot = 1
    else:
        bobot = 0
    
    total_nilai += bobot * 3   # karena setiap mata kuliah 3 SKS

total_sks = jumlah_mk * 3
ips = total_nilai / total_sks

print("IPS Anda adalah:", ips)
```
