```python
with open("file1.txt", "w") as f1:
    f1.write("Halo Dunia\n")
    f1.write("Belajar Python\n")
    f1.write("Saya suka coding\n")

with open("file2.txt", "w") as f2:
    f2.write("Halo Dunia\n")
    f2.write("Belajar Jupyter\n")
    f2.write("Saya suka coding\n")

try:
    file1 = open("file1.txt", "r")
    file2 = open("file2.txt", "r")
    
    baris1 = file1.readlines()
    baris2 = file2.readlines()
    
    maks = max(len(baris1), len(baris2))
    
    print("Hasil Perbandingan:\n")
    
    for i in range(maks):
        teks1 = baris1[i].strip() if i < len(baris1) else "(kosong)"
        teks2 = baris2[i].strip() if i < len(baris2) else "(kosong)"
        
        if teks1 != teks2:
            print("Baris ke-", i+1)
            print("File1 :", teks1)
            print("File2 :", teks2)
            print()
    
    file1.close()
    file2.close()

except FileNotFoundError:
    print("File tidak ditemukan!")
```

```python
with open("soal.txt", "w") as file:
    file.write("1+1 = || 2\n")
    file.write("Bendera Indonesia? || Merah Putih\n")
    file.write("Kota gudeg adalah: || Yogyakarta\n")
    file.write("Komponen PC untuk penyimpanan file adalah... || harddisk\n")
    file.write("50 * 20 = || 1000\n")

try:
    file = open("soal.txt", "r")
    print("Nama file:", file.name)
    print()

    for baris in file:
        data = baris.strip().split("||")
        soal = data[0].strip()
        jawaban_benar = data[1].strip()

        print(soal)
        jawaban = input("Jawab: ")

        if jawaban.lower() == jawaban_benar.lower():
            print("Jawaban benar!\n")
        else:
            print("Jawaban salah!\n")

    file.close()

except FileNotFoundError:
    print("File tidak ditemukan!")
```
